```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df=pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
mice_count = combined_df['Mouse ID'].nunique()
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df[combined_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicate_mice_ID=duplicate_mice['Mouse ID'].unique()
duplicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice['Mouse ID']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combined_df[combined_df['Mouse ID'] != 'g989']
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Bar and Pie Charts

```python
# calculate total rows per drug in new dataframe and add 'Total Row' Index
per_drug_df = clean_df.groupby('Drug Regimen').size().reset_index(name='Total Rows')

# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas
bar_plot=per_drug_df.plot(kind='bar', x='Drug Regimen', y='Total Rows', legend=False)
bar_plot.set_title('Drug Regimen Counts')
bar_plot.set_xlabel('Drug Regimen')
bar_plot.set_ylabel('Count')
bar_plot.figure.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
plt.bar(per_drug_df['Drug Regimen'], per_drug_df['Total Rows'])
plt.title('Drug Regimen Counts')
plt.xlabel('Drug Regimen')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
sex_df = clean_df.groupby('Sex').size().reset_index(name='Sex_Count')
pie_plot=sex_df.plot(kind='pie', y='Sex_Count', labels=sex_df['Sex'], autopct='%1.1f%%', legend=False)
pie_plot.set_title('Distribution of Male vs. Female')
pie_plot.figure.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(sex_df['Sex_Count'], labels=sex_df['Sex'], autopct='%1.1f%%')
plt.title('Distribution of Male vs. Female')
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
sorted_df = clean_df.sort_values(by=['Timepoint'], ascending=False)
last_tumor = clean_df.loc[clean_df.groupby('Mouse ID')['Timepoint'].idxmax()]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
Treatments = clean_df['Drug Regimen'].unique().tolist()

# Create empty list to fill with tumor vol data (for plotting)
Tumor_Vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in Treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_data = last_tumor[last_tumor['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    
    # add subset 
    Tumor_Vol.append(drug_data)

    
    # Determine outliers using upper and lower bounds
    lower_bound = drug_data.quantile(0.25) - 1.5 * st.iqr(drug_data)
    upper_bound = drug_data.quantile(0.75) + 1.5 * st.iqr(drug_data)
    outliers = drug_data[(drug_data < lower_bound) | (drug_data > upper_bound)]
                        
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(Tumor_Vol, labels=Treatments)
plt.title('Tumor Volume by Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Drug Regimen')
plt.xticks(rotation=90)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_data = clean_df[clean_df['Drug Regimen'] == 'Capomulin']
s185_data = capomulin_data[capomulin_data['Mouse ID'] == 's185']

plt.plot(s185_data['Timepoint'], s185_data['Tumor Volume (mm3)'], marker='o', linestyle='-', color='b')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Volume Over Time For Mouse s185 (Capomulin)')
plt.show()
```

```python
# Get Data for average tumor calc
Capomulin_Avg = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
Capomulin_merged = pd.merge(Capomulin_Avg, capomulin_data[['Mouse ID', 'Weight (g)']], on= 'Mouse ID', how='left')

# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
plt.scatter(Capomulin_merged['Weight (g)'], Capomulin_merged['Tumor Volume (mm3)'], color='b', alpha=0.7)
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight and Average Tumor Volume (Capomulin)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

Pearsonr_Capomulin = st.pearsonr(Capomulin_merged['Weight (g)'], Capomulin_merged['Tumor Volume (mm3)'])

#Calculate regression line
slope, intercept, r_value, p_value, std_error = st.linregress(Capomulin_merged['Weight (g)'], Capomulin_merged['Tumor Volume (mm3)'])

plt.scatter(Capomulin_merged['Weight (g)'], Capomulin_merged['Tumor Volume (mm3)'], color='b', alpha=0.7)
plt.plot(Capomulin_merged['Weight (g)'], slope * Capomulin_merged['Weight (g)'] + intercept, color='r')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight and Average Tumor Volume (Capomulin)')
plt.show()
print(f'The correlation coefficient for Mouse Weight and Average Tumor Volume for Mice on the Capomulin Regimen is {slope}')
```
